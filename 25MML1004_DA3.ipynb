# a) Structured Streaming in Spark

## Aim

To understand Structured Streaming in Spark, its features, differences from batch processing, and real-time applications.

## Introduction

Structured Streaming is a real-time data processing engine built on Spark SQL. It processes live data streams using the same API as batch processing.

## Concept of Structured Streaming

Structured Streaming treats streaming data as an unbounded table and processes it continuously in micro-batches.

## Difference Between Batch and Streaming

| Feature    | Batch Processing    | Structured Streaming |
| ---------- | ------------------- | -------------------- |
| Data Type  | Finite              | Infinite             |
| Processing | One-time            | Continuous           |
| Latency    | High                | Low                  |
| Use Case   | Historical analysis | Real-time analytics  |

## Working Flow

```
Data Source (Kafka / Socket / File)
            ↓
    Spark Streaming Engine
            ↓
    Transformation Logic
            ↓
  Output Sink (Console / Database)
```

## Key Features

* Unified API for batch and streaming
* Fault tolerance using checkpoints
* Scalability for large data streams
* Exactly-once processing guarantee
* Support for window operations

## Real-Time Use Cases

* Real-time analytics dashboards
* Fraud detection systems
* Log monitoring
* IoT data processing
* Social media analysis

## Example

```
df = spark.readStream.format("socket").load()

word_counts = df.groupBy("word").count()

query = word_counts.writeStream.format("console").start()
```

## Conclusion

Structured Streaming enables efficient real-time data processing using Spark. Its low latency, scalability, and fault tolerance make it suitable for modern data-driven applications.

````python
# Import Libraries

from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, split
import os
````

* SparkSession → Entry point for PySpark
* explode() → Converts array into multiple rows
* split() → Splits text into words

````python
# Create Spark Session

spark = SparkSession.builder \
    .appName("StructuredStreamingWordCount") \
    .getOrCreate()
````

* Initializes the Spark application

````python
# Create Streaming Input Directory

input_path = "/content/stream_data"
os.makedirs(input_path, exist_ok=True)
````

````python
# Read Streaming Data from Directory

lines = spark.readStream \
    .format("text") \
    .load(input_path)
````

Reads text files added to the folder as a stream
Each new file acts as incoming data

````python
# Perform Word Count Transformation

words = lines.select(
    explode(
        split(lines.value, " ")
    ).alias("word")
)

word_counts = words.groupBy("word").count()
````

split() → splits sentence into words
explode() → creates one row per word
groupBy() + count() → counts word frequency

````python
# Write Output to Console

query = word_counts.writeStream \
    .outputMode("complete") \
    .format("console") \
    .start()
````

````python
# Start Streaming

query.awaitTermination()
````
