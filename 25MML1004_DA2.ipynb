# a) Transformations vs Actions in Spark

## Aim

To understand the difference between transformations and actions in Spark and their role in execution planning.

## Introduction

In Apache Spark, operations are categorized into transformations and actions. These operations define how data is processed and when execution takes place.

## Transformations

Transformations are operations that create a new dataset from an existing one. They are lazy in nature, meaning they are not executed immediately.

Examples of transformations:

* map() – transforms each element
* filter() – filters rows based on condition
* select() – selects specific columns

Characteristics:

* Return new DataFrame or RDD
* Do not trigger execution
* Used to build logical execution plan

## Actions

Actions are operations that trigger execution and produce results.

Examples of actions:

* show() – displays data
* count() – returns number of rows
* collect() – returns all data to driver

Characteristics:

* Trigger execution
* Produce output
* Execute all prior transformations

## Execution Flow

```
Transformations → Transformations → Transformations
                      ↓
                   Action
                      ↓
             Execution Triggered
```

## Comparison Table

| Feature            | Transformations | Actions         |
| ------------------ | --------------- | --------------- |
| Execution          | Lazy            | Immediate       |
| Output             | New dataset     | Result          |
| Triggers Execution | No              | Yes             |
| Example            | filter(), map() | show(), count() |

## Significance in Execution Planning

Transformations help Spark build a Directed Acyclic Graph (DAG), which represents the sequence of operations. Execution is triggered only when an action is called. This approach allows Spark to optimize the execution plan and improve performance.

## Example

```
df.filter(col("age") > 25).select("name").show()
```

In this example:

* filter() and select() are transformations
* show() is an action

## Conclusion

Transformations and actions form the core of Spark’s execution model. Lazy evaluation allows Spark to optimize execution, making data processing efficient and scalable.

# b) PySpark Program: Data Transformations on Dataset

### Aim

The aim of this experiment is to use PySpark to:

* Load a dataset (CSV or manually created)
* Perform transformations such as selection, filtering, and grouping
* Display results using appropriate PySpark actions

### Objective

* To understand how to load data into PySpark DataFrame
* To perform transformation operations like:
  
  * Selecting specific columns

  * Filtering data
  
  * Grouping and aggregation
* To display results using actions like .show()

````python
# Install PySpark
!pip install pyspark
````

````python
# Import required modules
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count
````

````python
# Create Spark Session

spark = SparkSession.builder \
    .appName("PySpark Transformations Example") \
    .getOrCreate()
````

SparkSession is the entry point for all PySpark operations

````python
# Manually Create Dataset

data = [
    (1, "Alice", 25, "HR", 25000),
    (2, "Bob", 30, "IT", 40000),
    (3, "Charlie", 28, "IT", 32000),
    (4, "David", 35, "Finance", 28000),
    (5, "Eve", 29, "HR", 50000)
]

columns = ["ID", "Name", "Age", "Department", "Salary"]

df = spark.createDataFrame(data, columns)
````

````python
# Select Specific Columns

selected_df = df.select("Name", "Department", "Salary")
selected_df.show()
````


* select() is used to retrieve only required columns
* Reduces unnecessary data processing

````python
# Apply Filter Condition

filtered_df = df.filter(col("Salary") > 30000)
filtered_df.show()
````

* Filters employees with salary greater than 30,000
* col() helps reference column names

````python
# Group By and Count

grouped_df = df.groupBy("Department").agg(count("*").alias("Employee_Count"))
grouped_df.show()
````

* groupBy() groups data based on department
* count() counts number of employees in each group
* alias() renames the column

````python
# Sort by Salary

df.orderBy(col("Salary").desc()).show()
````

````python
# Display First Record

df.first()
````

````python
# Count Total Records

df.count()
````

### Conclusion

This experiment demonstrates how PySpark performs efficient data transformations on structured datasets. Using operations like selection, filtering, and grouping, we can analyze large datasets easily in a distributed environment.
