# a) Distributed Data Processing and Apache Spark Performance

## Aim

To understand the concept of distributed data processing and analyze how Apache Spark improves performance compared to traditional systems, including the role of in-memory computation.

## Introduction

Distributed data processing is a technique in which large datasets are divided into smaller parts and processed across multiple machines simultaneously. This approach is essential for handling big data, where a single system is insufficient to process large volumes of data efficiently.

## Concept of Distributed Data Processing

In distributed data processing:

* Data is split into multiple partitions
* Each partition is processed on different nodes
* Processing happens in parallel
* Results are combined to produce final output

Flow representation:

```
Large Dataset
      ↓
Split into Partitions
  ↓    ↓    ↓
Node1 Node2 Node3
  ↓    ↓    ↓
```

Parallel Processing
↓
Combined Output

## Traditional Data Processing Limitations

Traditional systems such as disk-based frameworks have the following drawbacks:

* Heavy dependency on disk I/O
* Slower execution due to repeated read/write operations
* Inefficient for iterative tasks
* Not suitable for real-time processing

## How Apache Spark Improves Performance

Apache Spark is a fast distributed processing engine designed to overcome these limitations. It improves performance through:

* In-memory computation
* Lazy evaluation
* DAG-based execution
* Parallel processing

### In-Memory Computation

In-memory computation allows Spark to store intermediate data in RAM instead of disk. This reduces I/O operations and significantly speeds up processing, especially for iterative tasks such as machine learning.

### Comparison Table

| Feature              | Traditional Systems | Apache Spark    |
| -------------------- | ------------------- | --------------- |
| Storage              | Disk-based          | In-memory (RAM) |
| Speed                | Slow                | Fast            |
| Iterative Processing | Inefficient         | Efficient       |
| Real-time Capability | Limited             | Strong          |

## Conclusion

Distributed data processing enables efficient handling of large datasets by using multiple systems. Apache Spark enhances performance through in-memory computation, parallel processing, and optimized execution strategies, making it significantly faster than traditional systems.

# b) PySpark Program: Employee DataFrame Operations

### Aim

The aim of this experiment is to demonstrate how to use PySpark to:

* Create a DataFrame with employee records
* Define a schema (ID, Name, Age, Salary)
* Display the structure (schema) of the DataFrame
* Show all records present in the DataFrame
* Filter employees based on a salary condition (Salary > 30,000)

### Objective
* To understand how PySpark handles structured data
* To perform basic DataFrame operations such as creation, visualization, and filtering
* To learn schema inspection and transformation techniques

````python
!pip install pyspark
````

````python
# Import required libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
````

````python
# Create Spark Session
spark = SparkSession.builder \
    .appName("Employee DataFrame Example") \
    .getOrCreate()
````



* SparkSession is the entry point to PySpark
* It allows us to create DataFrames and execute operations

````python
# Create DataFrame
# Sample data
data = [
    (1, "Alice", 25, 25000),
    (2, "Bob", 30, 40000),
    (3, "Charlie", 28, 32000),
    (4, "David", 35, 28000),
    (5, "Eve", 29, 50000)
]

# Column names
columns = ["ID", "Name", "Age", "Salary"]

# Create DataFrame
df = spark.createDataFrame(data, columns)
````


* Data is created as a list of tuples
* Columns define schema
* DataFrame is created using createDataFrame()

````python
# Display Schema
# Print schema
df.printSchema()
````

* Shows structure of DataFrame
* Displays column names, data types, and nullability

````python
# Show all records
df.show()
````

* Displays all rows in tabular format
* Useful for quick inspection of data

````python
# Filter Employees with Salary > 30,000
# Filter employees

filtered_df = df.filter(col("Salary") > 30000)

# Show filtered results
filtered_df.show()
````

* filter() is used to apply condition
* col("Salary") > 30000 selects only high-salary employees

````python
# Select Specific Columns

df.select("Name", "Salary").show()
````

````python
# Sort by Salary

df.orderBy(col("Salary").desc()).show()
````

````python
# Add New Column

df.withColumn("Bonus", col("Salary") * 0.1).show()
````

###Conclusion

This experiment demonstrates how PySpark simplifies big data processing using DataFrames. It provides efficient operations such as filtering, transformation, and schema handling, making it a powerful tool for distributed data processing.
